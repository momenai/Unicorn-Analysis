```python
max_valuation_index = unicorns["Valuation ($B)"].idxmax()
highest_valuation = unicorns.loc[max_valuation_index, "Valuation ($B)"]
company_name = unicorns.loc[max_valuation_index, "Company"]
print(f"The highest valuation of a unicorn company is {highest_valuation} billion dollars, and the company name is {company_name}.")
```

```python
top_two = unicorns.nlargest(2, "Valuation ($B)")
second_highest_valuation = top_two.iloc[1]["Valuation ($B)"]
company_name = top_two.iloc[1]["Company"]
print(f"The second highest valuation of a unicorn company is {second_highest_valuation} billion dollars, and the company name is {company_name}.")
```
